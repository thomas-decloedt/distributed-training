# Distributed Training Concepts

**Data parallelism**: Same model on each worker; each processes different data batches. Gradients are synchronized so all workers update consistently.

**Model parallelism**: Different layers on different devices. Used for models too large to fit on one GPU.

**PyTorch DDP**: Uses `torch.distributed` to synchronize gradients via all-reduce during the backward pass.
